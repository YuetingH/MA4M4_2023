## Read a file and create a network from its edgelist

You can find and download the Zachary karate club dataset [here](http://konect.cc/networks/ucidata-zachary/).

Have the download documents ready in the same folder of this jupyter notebook (you can have it somewhere else, just pass the path of the files when inputting the data file)

```python
fh = open("out.ucidata-zachary", "rb")  # 'rb': open the file in binary format for reading
G = nx.read_edgelist(fh, comments='%',create_using=nx.DiGraph())   
    # comments='%': the character used to indicate the start of a comment.
    # create_using=nx.DiGraph(): create a directed graph allowing self loops but not multiple edges.(default=nx.Graph)
fh.close()
```

## Input the edgelist as a Pandas dataframe

Sometimes we may consider processing the edges before building the network. Pandas dataframe can be helpful in this case. Pandas is a handy tool to process data and is widely used not only in network science but also in other data science related fields.

In this example, we look at a small directed weighted graph ([kangaroo](http://konect.cc/networks/moreno_kangaroo/)) (with visualisation purpose), which is a bit more complicated than an undirected and unweighted case. 

```python
df = pd.read_csv("out.moreno_kangaroo_kangaroo",sep=' ',skiprows=[0])
    # sep=' ': separator(default type is ',')
    # skiprows=[0]: skip the first line
cols = [0,1,2]
df = df[df.columns[cols]] # return the first three columns
    # df.columns: return index of columns(use name to represent index)
df.columns=['from','to','weight'] # rename columns
df.head() # viewing the first 5 lines
```

```python
df.describe()
```

Then, we import this network from pandas directly.

```python
# notice how input values consistent with pandas column names.
G = nx.from_pandas_edgelist(df,source='from',target='to',edge_attr='weight',create_using=nx.DiGraph)
    # edge_attr is where we state the weight information
    # create_using=nx.DiGraph(): create a directed graph allowing self loops but not multiple edges.
```

```python
nx.info(G)
```

```python
G.nodes
```

```python
G.edges
```

```python
nx.get_edge_attributes(G, 'weight')
```

## Create a network from its adjacency matrix

We can also build the network by inputting an adjacent matrix.

Firstly, let's compute an adjacency matrix from the edgelist and visualise it.

```python
adj0 = nx.adjacency_matrix(G, nodelist=None, dtype=None, weight='weight')
```

Note that adj0 is of sparse matrix type. Some usual numpy operations on dense matrices may not work. For example, if you want to read it, you need to turn it into a dense type.

```python
adj0 # A sparse matrix is a matrix that is comprised of mostly zero values
```

Think about why it is an upper triangular matrix?

```python
adj0.todense() # turn into a dense type (return a matrix)
```

Let's turn this into an unweighted directed matrix. We are going to build a network using the new matrix.

```python
unweighted_adj = adj0.copy() # so that change in unweighted_adj won't affect adj0
unweighted_adj[unweighted_adj>0] = 1

# one way to visualise a sparse matrix
print(unweighted_adj.todense())

# another way to visualise a sparse matrix
plt.figure(figsize=(5,5))
plt.spy(unweighted_adj)
```

```python
G_unweighted = nx.DiGraph(unweighted_adj)
G_unweighted.edges.data()
```

Notice we still have the 'weight' feature though they all have value 1. We can delete this feature if we want.

```python
for u,v in G_unweighted.edges:
    # u: starting point of edges
    # v: end point of edges
    del G_unweighted[u][v]['weight']
```

```python
nx.is_weighted(G_unweighted)
```

```python
nx.draw(G_unweighted, pos = nx.spring_layout(G_unweighted, k=5), with_labels=True) # Take times for large network
    # k helps determine the distance between nodes
plt.axis("on");
```

## Get the degree sequence of all nodes and plot its histogram

```python
degree= np.array([b for x,b in nx.degree(G_powergrid)])
    # nx.degree(G_powergrid): return a list of (node, degree) pairs
print(len(degree))
```

```python
pk = degree/np.sum(degree)
```

```python

plt.hist(degree,density=True)
```

##  Plot the degree distribution in loglog scale. Can you observe power law in the plot? Think about how you can approximate the α.

1. Start by finding P(x=k) from degree sequence
2. Plot P(x=k)

```python
degree_freq = np.bincount(degree)
degrees = range(len(degree_freq))
```

```python
plt.figure(figsize=(10, 5)) 
plt.scatter(degrees, degree_freq/np.sum(degree_freq)) 
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Degree')
plt.ylabel('Frequency')
```

## Using the P(x=k) you find above, try plotting its cdf i,e P(x>=k), also in log log scale

```python
plt.figure(figsize=(10,5))
plt.step(degrees,1-np.cumsum(degree_freq/np.sum(degree_freq)),'co-') # or simply plt.loglog. Try it out.
plt.grid()
plt.xscale('log')
plt.yscale('log')
plt.ylabel("CDF")
plt.xlabel("k")
```
